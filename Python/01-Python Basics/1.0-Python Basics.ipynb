### Python is case sensitive

```python
##### Variable is a container that stores the information. Python is case sensitive so both the variables are different.
name="Sowmi"
Name="P"

print(name)
print(Name)
```

```python
## Indentation
## Python uses indentation to define blocks of code. Consistent use of spaces (commonly 4) or a tab is required.

age=25
if age < 30:
    print(age)
else:
    print(age)


print("HELLO WORLD")
```

```python
## Line Continuation:: Use a backslash (\) to continue a statement to the next line
total=1+2+3+4+5+6+7+\
4+5+6
print(total)

## Multiple Statements on a single line
x=5;y=10;z=x+y
print(z)
```

```python
#Semnatics In Python (variable assignment)
age=32 ##age is an integer
name="Krish" ##name is a string

print(type(age))
print(type(name))
```

```python
## Type Inference:: # The type of variable changed from integer to string based on the assigned value. During the runtime(Dynamic Typing), we can chane the type of variable.
variable=10
print(type(variable))
variable="Krish"
print(type(variable))
```
