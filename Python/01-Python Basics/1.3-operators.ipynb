##### Operators

```python
## Arithmethic Operation

a=10
b = 5

add_result = a+b  #addiiton
sub_result = a-b  #substraction 
mult_result = a*b #multiplication
div_result = a/b  #division
floor_div_result = a//b ## floor division
modulus_result = a%b #modulus operation: gives the remainder
exponent_result = a**b ## Exponentiation


print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exponent_result)
```

#### Comparison Operators

##### Logical Operators: AND, OR, NOT

```python
# AND: IF A AND B IS TRUE THEN RESULT IS TRUE; IF NOT THEN FALSE
a = 5
b = 10
print( a < 10 and b > 5)

print(a < 10 and b < 5)

## And 
X=True
Y=True
result = X and Y
print(result)

X=False
Y=True
result = X and Y
print(result)
```

```python
## OR: IF A OR B IS FALSE THEN RESULT IS FALSE; IF NOT THEN TRUE
X=False
Y=False

result = X or Y
print(result)
```

```python
# Not operator
X= False
not X
```
