##### In Python, data types define the kind of value a variable can hold and determine what operations can be performed on that value. Since everything in Python is an object, each data type is actually a class, and variables are instances (objects) of these classes

#### DataTypes:
###### 1. Definition: Data types are a classification of data which tell the compiler or interpreter how the programmer intends to use the data.
###### 2. Importance of Data Types in Programming is Data types ensure that data is stored in an efficient way. They help in performing correct operations on data. Proper use of data types can prevent errors and bugs in the program.

```python
#int examples
a = 5
print(type(a))

# float example
b = 3.15
print(type(b))

#string example
name = 'SOW'
print(type(name))

#boolean example
bool1 = True
print(type(bool1))
```

```python
a = 1
b = 5
print(type(a == b))
print(type(a != b))
```
