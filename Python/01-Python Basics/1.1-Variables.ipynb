##### Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable. Variables are essential in Python programming for storing and manipulating data. Understanding how to declare, assign, and use variables effectively is crucial for writing functional and efficient code. Following proper naming conventions and understanding variable types will help in maintaining readability and consistency in your code.

```python
# Declaring and Declaring Variables
age = 26
name = "Sow"
height = 5.6
is_Student = True

print("age:", age)
print("name:", name)
print("height:", height)
```

```python
## Naming Conventions
## Variable names should be descriptive. They must start with a letter or an '_' and contains letter,numbers and underscores and variables names case sensitive

## case sensitivity, these two variables are different
name="SOW"
Name="SOW"

#Valid conventions
first_name="Sow"
last_name="PUL"

#Invalid conventions
#1st_name="Sow" (starts with a number)
#first-name = 'SOW' (contains hyphen)
#@name = 'Sow' (starts with special character)
```

```python
## Understanding Variable types
## Python is dynamically typed, meaning the type of a variable is determined at runtime
age = 26
name = "Sow"
height = 5.6
is_Student = True
print(type(age))     # These functions return the type of the variable, this is type checking
print(type(name))
print(type(height))
print(type(is_Student))
```

```python
#Type Conversion
# int to str
age = 26
print(type(age))
age_str = str(age)
print(type(age_str))

# str to int
age = '25'
print(type(age))
print(type(int(age)))

# str with sentence to int
name = "SOW"
print(type(int(name)))
```

```python
height = 5.11
print(type(height)) # checking type
print(int(height))  # float to int
```

```python
# Dynamic Typing
### Python allows the type of a varible to change as the program executes
var = 10 #integer
print(var,type(var))

var = "sow" # string for the same variable
print(var,type(var))

var = 3.11
print(var,type(var))
```
