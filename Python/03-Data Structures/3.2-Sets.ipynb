##### Sets: Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

```python
# create a set
set = {1,2,3,5,4,6}
print(set)
print(type(set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
my_set = set([1,2,3,4,5,6])
print(my_set)
```

```python
## Set Membership test
my_set = {1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operations
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

###Union 
union1 = set1.union(set2)
print(union1)


## Intersection
intersection1 = set1.intersection(set2)
print(intersection1)

## Intersection update ::: all the intersection elements will update in the set1
set1.intersection_update(set2)
print(set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}


##Difference
print("set2_diff:", set1.difference(set2))

print("set2_diff:", set2.difference(set1))

print(set1)
print(set2)
```

```python
## Symmetric Difference:::: unique elements from both the sets are combined (common elements from the sets are removed)
set1.symmetric_difference(set2)
```

```python
### Sets Methods
set1 = {1,2,3}
set2 = {3,4,5}

## issubset(): Checks if all elements of one set are present in another set
print(set1.issubset(set2))


### issuperset(): Checks if a set contains all elements of another set
print(set1.issuperset(set2))
```

```python
lst = [1,2,3,5,6,5,6,6]

## to remove the duplicates from the list; then convert it into the set
set(lst)
```
