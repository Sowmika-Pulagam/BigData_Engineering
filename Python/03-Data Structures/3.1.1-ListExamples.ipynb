```python
##### Example 1. Manage A To Do List: Create a To Do List To Keep Track OF Tasks
to_do_list=["Buy Groceries","Clean the house","Pay bills"]

# Adding to task
to_do_list.append("schedule doctor appointment")
to_do_list.append("Go for a run")

# Removing a task
to_do_list.remove("Clean the house")

# Checking the task is in the list
if "pay bills" in to_do_list:
    print("Don't forgot to pay the bills")

print("To Do List:")
for task in to_do_list:
    print(task)
```

```python
##### Example 2: Organizing Student Grades: Create a list to store and calculate average grades for students
grades = [85, 92, 78, 90, 88]

#adding a new grade
grades.append(95)

## Calculating the average grade
average_grade = sum(grades)/len(grades)
print(f"Average Grade: {average_grade}")

highest_grades = max(grades)
lowest_grades = min(grades)
print(f"Highest Grade: {highest_grades}")
print(f"Lowest Grade: {lowest_grades}")
```

```python
##### Example 3: Managing An Inventory: Use a list to manage inventory items in a store
inventory = ["Apples", "Bananas", "Oranges", "Grapes"]

# Adding new items to inventory
inventory.append("watermelon")

# Removing an item that is out of stock
item = "oranges"
if item in inventory:
    print(f"{item} is in stock.")
else:
    print(f"{item} is out of stock")


# Printing the inventory
print("Inventory List:")
for item in inventory:
    print(item)

```

```python
##### Example 4:Collecting User Feedback: Use a list to collect and analyze user feedback
feedback = ["Great service!", "Very satisfied", "Could be better", "Excellent experience"]

# Adding new feedback
feedback.append("Improve the cleanliness")

# Analyzing feedback
print("User Feedback:")
for comment in feedback:
    print(comment)
```
