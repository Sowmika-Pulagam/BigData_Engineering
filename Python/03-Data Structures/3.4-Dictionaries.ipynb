#### Dictionaries
Dictionaries are unordered collections of items. They store data in key-value pairs. Dictionary are mutable. Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
## Creating Dictionaries
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
print(empty_dict)
```

```python
student = {'name': "krish", "age": "26", "grade": 24}
print(student)
```

```python
# the one with the recent value will be updated in the key name instead of krish we get 24 
student = {'name': "krish", "age": 26, "name": 24}
print(student)
```

```python
### Accessing Dictionary elements:
student = {'name': "krish", "age": 26, "grade": 'A'}
print(student['grade'])

print(student['name'])
```

```python
## Accessing using get() method
print(student.get('age'))

print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
## Modifying Dicitonary Elements
## Dictionary are mutable,so you can add, update or delete elements but key needs to be unique all the time
print(student)
```

```python
##update value for the key
student["age"] = 33
print(student)

## added a new key and value
student["address"] = "USA"
print(student)
```

```python
# deleted a key and value pair
del student["age"]
print(student)
```

```python
## Dictionary methods
## TO get only the keys in dict
print(student.keys())
# To get only the values in dict
print(student.values())

## To get all the key-value pairs
print(student.items())
```

```python
## Why we do Shallow Copy
student_copy = student
print(student)
print(student_copy)
```

```python
student["name"] = "Krish2"
print(student)
print(student_copy)
```

```python
## Shallow Copy: provides different memory location, when we changes a value of varibale another data should not be affected

student_copy1 = student.copy()
print(student_copy1)
print(student)
```

```python
student["name"] = "Krish3"
print(student_copy1)
print(student)
```

```python
### Iterating Over Dictionaries
### You can use loops to iterate over dictionatries (keys, values, items)

# Iterating over key in student:
 
for key in student.keys():
    print(key)
```

```python
# Iterating over values in student:
for value in student.values():
    print(value)
```

```python
# Iterating over key-value in student:
for key, value in student.items():
    print(f"{key} : {value}")
```

```python
## Nested Dictionaries

students = {
    "student1": {"name": "Sow", "age": 26},
    "student2": {"name": "Vee", "age": 32}
}
print(students)
```

```python
## Access nested dictionary elements
print(students["student2"]["name"])
print(students["student1"]["age"])
```

```python
students.items()
```

```python
## Iterating over nested dictionaries

for student_id, student_info in students.items():
    print(f"{student_id} : {student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")

```

```python
## Dictionary Comphrehension

squares = {x: x ** 2 for x in range(5)}
print(squares)
```

```python
## Condition dictionary comprehension

evens = {x: x **2 for x in range(10) if x % 2 == 0}
print(evens)
```

```python
## Practical Examples
## Use a dictionary to count the frequency of elements in list

numbers = [1,2,2,3,3,3,4,4,4,4]
frequency = {}

for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

print(frequency)
```

```python
## Merge 2 dictionaries into one
dict1 = {"a": 1 , "b": 2}
dict2 = {"b": 3 , "c": 4}
merged_dict = { **dict1, **dict2} # unpacking the dictionary
print(merged_dict)
```
