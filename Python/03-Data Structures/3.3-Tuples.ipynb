#### Introduction to Tuples

#### Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.

```python
### Creating a tuple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))

tpl = tuple()
print(type(tpl))
```

```python
## TYPE CONVERSIONS
### Converting a list into tuple
tuple_numbers = tuple([1,2,3,4,5,6])
print("tuple:",tuple_numbers)

### Converting a tuple into list
list_numbers = list((1,2,3,4,5,6))
print("list:",list_numbers)
```

```python
## Combination of types
mixed_tuple=(1,"Hello World",3.14, True)
print(mixed_tuple)
```

```python
### Accessing Tuple Elements
numbers = (1,2,3,4,5,6)

print(numbers[0])
print(numbers[2])

print(numbers[-1])

# slicing
print(numbers[1:3])
## Reverse a tuple
print(numbers[::-1])
```

```python
### Tuple Operations
concatenation_tuples = numbers + mixed_tuple
print(concatenation_tuples)
```

```python
mixed_tuple * 2
```

```python
numbers * 2
```

```python
## Immutable Nature Of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned. # Example of tuple


## List are muttable meaning, they can be changed after assigning ## Example of list
lst=[1,2,3,4,5]
print(lst)

lst[1]="Krish"
print(lst)
```

```python
numbers[1] = 0
```

```python
## Tuple Methods ## 2methods we specifically use: count and index
print(numbers.count(1)) ## this checks how many number of time the element is present in the tuple

print(numbers.index(3)) ## which index the number 3 is present
```

```python
## Packing and Unpacking tuple

## packing a tuple
packing_tuple = 1,"Hello",3.14
print(packing_tuple)

## Unpacking a tuple
a, b, c = packing_tuple
print(a) 
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)

first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
#### Nested

##Nested List (Inside a list we can have tuples)
lst = [[1,2,3,4],[2,4,3,5,6],[1,"Hello",3.14,"c"]]
print(lst[0])
print(lst[0][2])
print(lst[0][0:3])
```

```python
#### Nested list with tuples
lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
print(lst[2][0:3])
```

```python
#### Nested TUPLES
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

#### Conclusion
Tuples are versatile and useful in many real-world scenarios where an immutable and ordered collection of items is required. They are commonly used in data structures, function arguments and return values, and as dictionary keys. Understanding how to leverage tuples effectively can improve the efficiency and readability of your Python code.
