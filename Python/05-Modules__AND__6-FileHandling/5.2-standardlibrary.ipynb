##### Standard Library Overview
Python's Standard Library is a vast collection of modules and packages that come bundled with Python, providing a wide range of functionalities out of the box. Here's an overview of some of the most commonly used modules and packages in the Python Standard Library.

```python
import array
arr = array.array('i', [1, 3, 78])
print(arr)
```

```python
import math
print(math.sqrt(20))
print(math.pi)
```

```python
## random 
import random
print(random.randint(1,20))
print(random.choice(['apple', 'cherry', 'orange', 'blueberries']))
```

```python
### File And Directory Access
import os
print(os.getcwd())
```

```python
os.mkdir('test_dir')  # function in Python is used to create a new directory at a specified path
```

```python
## High level operations on files and collection of files
import shutil
shutil.copyfile('source.txt','destination.txt') 
### Copies the contents of the file source.txt to a new file named destination.txt AND Overwrites destination.txt if it already exists.
```

```python
## Data Serialization::: process of converting Python objects into a format that can be stored (e.g., in a file or database) or transmitted (e.g., over a network) 
# and later reconstructed (deserialized) back into the original object. Common serialization formats include JSON, Pickle, YAML, and MessagePack.

import json
data={'name':'Krish','age':25}

json_str = json.dumps(data) ## Serialize obj to a JSON formatted str
print(json_str)
print(type(json_str))


json_json = json.loads(json_str) # Deserialize s (a str, bytes or bytearray instancecontaining a JSON document) to a Python object.
print(json_json)
print(type(json_json))
```

```python
## csv

import csv

with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Krish',32])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
## datetime
from datetime import datetime,timedelta

now = datetime.now()
print(now)

yesterday= now- timedelta(days=1)  # timedelta(days=1) → Represents a time duration of exactly 1 day.
print(yesterday)
```

```python
# time
import time
print(time.time())

time.sleep(2)
print(time.time())
```

```python
## Regular expresiion
import re

pattern = r'\d+' # looking for the digits
text = 'There are 123 apples 456' # 123 is matched so it stops looking
match = re.search(pattern,text)
print(match.group())
```
