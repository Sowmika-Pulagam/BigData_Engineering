#### File Operation- Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary.

```python
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# Read a file line by line
with open('example.txt', 'r') as file: # r is read
    for line in file:
        print(line.strip()) ## strip() removes the newline character)
```

```python
## Writing a file(Overwriting)
with open('example.txt', 'w') as file: # w is write
    file.write("Hi, there!\n")
    file.write("Hi, i am doing okay!")
```

```python
## Write a file(wwithout Overwriting)

with open('example.txt', 'a') as file:
    file.write('/nI have a coding class!\n')
```

```python
#### Writing a list of lines to a file 

lst = ['apple \n', 'banana \n', 'orange\n']
with open('example.txt', 'a') as file:  # a is append
    file.writelines(lst)
```

```python
### Binary Files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:  ## wb is writebinary
    file.write(data)
```

```python
# Reading a binary file
with open('example.bin', 'rb') as file: ## rb is readbinary
    x = file.read()
    print(x)
```

```python
### Read the content from a source text file and write to a destination text file
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
#Read a text file and count the number of lines, words, and characters.
# Counting lines, words, and characters in a text file
def count_text(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines, words, characters = count_text(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
```

The w+ mode in Python is used to open a file for both reading and writing. If the file does not exist, it will be created. If the file exists, its content is truncated (i.e., the file is overwritten).

```python
### Writing and then reading a file

with open('example.txt', 'w+') as file:
    file.write("Hello world\n")
    file.write("This is a new line \n")


    ## Move the file cursor to the beginning
    file.seek(0)  ## goes the cursor to the 0th line

    ## Read the content of the file
    content = file.read()
    print(content)
```
