#### Functions in Python

##### Introduction to Functions
A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.

```python
def function_name(parameters):
    """DocString""" # comments
    #Function body
    return expression
```

```python
def even_or_ordd(num):
    """This function finds even or odd"""
    if num % 2 == 0:
        print("num is even")
    else:
        print("num is odd")

#Calling the above function with function_name (even_or_ordd)
even_or_ordd(5)
even_or_ordd(10)
```

```python
### function with multiple parameters

def num(a,b):
    """this is add function"""
    c = a+b
    return c


# call the function
print(num(2,8))
print(num(10,9))
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name}!, welcome")

# call the function
greet("Sow")
```

```python
## Default Parameters

def greet(name = "Vee"): # This is default parameter
    print(f"Hello {name}!, welcome")

# call the function
greet("Sow") # if don't give the parameter inside the greet() function then i get default as "Vee"; if i give the parameter then that value
```

```python
### Variable Length Arguments::: Positional And Keywords arguments

## Positional arguments
def print_numbers(*args): # This is a positional argument (*args);  args meaning argumental position with single *
    for numbers in args:
        print(numbers)

# call the function
print_numbers(1,2,3,4,5,6,7,8,"Krish")
```

```python
## Keywords Arguments 
def print_details(**kwargs): # kwargs meaning keyword argumental with 2** and the value will be in the form of key,value pairs
    for key, value in kwargs.items():
        print(f"{key}: {value}")


# call the function
print_details(name = "Sow", age = 26, location = "US")
```

```python
# combine both positional and keyword arguments:
def print_details(*args, **kwargs):
    for val in args:
        print(f"positional argument: {val}")
    
    for key, value in kwargs.items():
            print(f"keyword argument: {key}: {value}")


# Call the function
print_details(1,2,3,4,5,6,7,8,"Krish", name = "Sow", age = 26, location = "US") ## always first write the positional arguments, then write the keyword argument
```

```python
### Return statements
def multiply(a,b,c):
    return a * b * c

# call the function
multiply(1,4,6)
```

```python
### can return multiple statements
def multiply(a,b,c):
    return a * b * c, b

# call the function
multiply(1,4,6)
```
