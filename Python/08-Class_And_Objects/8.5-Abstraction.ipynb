#### Abstraction
Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort.

```python
from abc import ABC,abstractmethod


## Abstract base class
class Vehicle(ABC): ## for the abstract class and normal class is that we need to write ABC for abstract class. ABC stands for: Abstract Base Class
# When you inherit from ABC, you are telling Python: "This class is meant to be a blueprint, not a fully usable class." Without ABC, Python will NOT treat your class as abstract.
    def drive(self):  ### This is a normal method (MEthods are nothing but a function inside a class)
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self): ## because this is a abstact method we need to use this in respective child classes car and MotorCycle
        pass

#child class1
class car(Vehicle):
    def start_engine(self): ## so this a astarctclass so we can't inherit here so writing it here; drive fro above  is automatically inherited
        print("Car engine is working")
    
class MotorCycle(Vehicle):
    def start_engine(self):
        print("Motorcycle engine started")


def operate_vehicle(vehicle):
    vehicle.drive()
    vehicle.start_engine()


car1 = car()
operate_vehicle(car1)

motor = MotorCycle()
operate_vehicle(motor)
```
