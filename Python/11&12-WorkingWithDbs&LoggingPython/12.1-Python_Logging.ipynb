#### Python Logging
Logging is a crucial aspect of any application, providing a way to track events, errors, and operational information. Python's built-in logging module offers a flexible framework for emitting log messages from Python programs. In this lesson, we will cover the basics of logging, including how to configure logging, log levels, and best practices for using logging in Python applications.

In Python, logging is the process of recording messages about a program’s execution so you can monitor, debug, and analyze its behavior. Python provides a built-in logging module that lets you record events with different severity levels (e.g., debug info, warnings, errors) and send them to various outputs like the console, files, or external systems.

Logging Levels
From lowest to highest severity:

1. DEBUG –  Detailed information, typically of interest only when diagnosing problems.
2. INFO – General events confirming things are working. Confirmation that things are working as expected.
3. WARNING – An indication that something unexpected happened or indicative of some problem in the near future. The software is still working as expected.
4. ERROR – A serious problem that prevented part of the program from working. Due to a more serious problem, the software has not been able to perform some function.
5. CRITICAL – A severe error that may cause the program to stop. A very serious error, indicating that the program itself may be unable to continue running.

```python
import logging
```

```python
## Configure the basic logging settings

logging.basicConfig(level = logging.DEBUG)

## log messages for debug
logging.debug("This is a debug message")

# for info
logging.info("This is a info message")

# for warning
logging.warning("this is a warning message")

# for error
logging.error("this is a error message")

# for critical
logging.critical("this is a critical message")
```

```python
## configuring logging
import logging

logging.basicConfig(
    filename = 'app.log',
    filemode = 'w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
    )

## log messages with different severity levels
## log messages for debug
logging.debug("This is a debug message")
# for info
logging.info("This is a info message")
# for warning
logging.warning("this is a warning message")
# for error
logging.error("this is a error message")
# for critical
logging.critical("this is a critical message")
```
