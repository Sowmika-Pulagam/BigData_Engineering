#### Logging with Multiple Loggers
You can create multiple loggers for different parts of your application.

In python logging module, there’s an important distinction between 
1. logger objects (ex: logger = logging.getLogger(...)) 
2. module-level logging functions (ex: logging.info(...)).

```python
import logging

# create a logger for a specific module1
logger1 = logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)


# create a logger for a specific module2
logger2 = logging.getLogger("module2")
logger2.setLevel(logging.WARNING)
```

```python
logging.basicConfig(
    level = logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
# log messages with different loggers
logger1.debug("This is a debug message for module1")
logger2.warning("This is a warning message for module2")
logger2.error("This is a error message for module2")
```
