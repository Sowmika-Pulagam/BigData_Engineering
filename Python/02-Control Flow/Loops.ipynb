#### Loops: for, while, loop control statements(break, continue, pass)

```python
# for loop: iterate over sequence of numbers
for i in range(5):
    print(i)


# for
for i in range(1,6):
    print(i)    
```

```python
# range with step
for i in range(1, 10, 2):
    print(i)


#range with reverse step
for i in range(10, 1 , -2):
    print(i)
```

```python
# strings
name = 'HELLO'
for i in name:
    print(i)
```

```python
# while loop: continues to execute as long as condition is true
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# Loop control statements
# break statement: break exists the loop immediately
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue: continue skips the current iteration and moves to the next iteration
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# pass: is a null statement, it does nothing
for i in range(5):
    if i == 3:
        print("Three found",i)
        pass
    print(i)
```

```python
# Nested loops (loop inside another loop)
for i in range(3):
    for j in range(2):
        print(f"i:{i}, j:{j}")
```

```python
## Examples- Calculate the sum of first N natural numbers using a while and for loop
n = 10
sum = 0
count = 1

# Using while loop
while count <= n:
    sum = sum + count
    count += 1

print(f"Sum of first 10 natural numbers using while loop is: {sum}")


# Using for loop
sum = 0
for i in range(11):
    sum = sum + i
print(sum)
```

```python
## Example- Prime numbers between 1 and 100
for num in range(1, 101):
    if num > 1:
        for i in range(2, int(num**0.5) + 1):
            if (num % i) == 0:
                break
        else:
            print(num)
```
