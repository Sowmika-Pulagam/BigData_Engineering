#### 1. above did the uncaching and again did the caching of the customers_df data. So, the customer_id_filtered.count() takes lot time becasue it has to do the caching and spark will go and look for each and every partition bcause it's a count

```python
customer_id_filtered = customers_df.filter((customers_df.customer_id>='100000') & (customers_df.customer_id<='4000000'))
```

```python
#### taking time because count has to go to all partitions; spark has to go to each and every partition and do the count 
customer_id_filtered.count()
```

```python
#### very fast because it's shows on;y 1partition top records
customer_id_filtered.show(5)
```

#### 2. Below code is very fast because it directly goes to the caching and does the filter and do the show action, where in show it only takes the 1partition and give the output

```python
filtered_df = customers_df.filter(customers_df.city=='Mumbai')
filtered_df.show(5)
```

#### 3. Below query going to take take to execute, groupBy is a wide transformation meaning combines all the data from the all the partitions and do the groupBy transformation

```python
agg_df = customers_df.groupBy('city').count()
agg_df.show()
```

```python
## going to take time to do the uncache because it's going to take time delete the data from the disk:
customers_df.unpersist()
```
