```python

from pyspark.sql.types import StructType, StructField, IntegerType, StringType, BooleanType, DateType

schema = StructType([
    StructField("customer_id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("city", StringType(), True),
    StructField("state", StringType(), True),
    StructField("country", StringType(), True),
    StructField("registration_date", DateType(), True),  # Enforcing Date Type
    StructField("is_active", BooleanType(), True)  # Enforcing Boolean Type
])
```

### FAILFAST: Stops processing immediately when a malformed record is encountered.

```python
hdfs_file_path = '/data/customers_with_errors.csv'

df_failfast = spark.read \
.format('csv')\
.option('header',"true")\
.schema(schema)\
.option('mode','FAILFAST')\
.load(hdfs_file_path)

df_failfast.show()
```

### PERMISSIVE (DEFAULT) :  Spark sets invalid fields to null.

```python
df_permissive = spark.read \
.format('csv')\
.option('header',"true")\
.schema(schema)\
.option('mode','PERMISSIVE')\
.load(hdfs_file_path)

df_permissive.show()
```

### DROPMALFORMED:Automatically remove malformed records.

```python
df_dm = spark.read \
.format('csv')\
.option('header',"true")\
.schema(schema)\
.option('mode','DROPMALFORMED')\
.load(hdfs_file_path)

df_dm.show()
```
