```python
from pyspark.sql import SparkSession
import spark
```

```python
### I have uploaded the customers_300mb.csv file into the local and kept it in the hadoop files 
### And reading the file from HDFS
!hadoop fs -ls /data/
```

```python
spark = SparkSession.builder \
.appName("Read Data") \
.getOrCreate()
```

```python
df = spark.read \
    .format("csv") \
    .option("header", "true") \ ## here we are not using inferSchema as true; we get 1task; 1partition
    .load("/data/customers_300mb.csv")
```

```python
# Read the file from HDFS
df = spark.read \
.format("csv") \
.option("header", "true") \
.option("inferSchema", "true") \ # here we are using inferSchema as true; we get 2task; 2partition
.load("/data/customers_300mb.csv")
```

```python
from pyspark.sql.types import *

schema = StructType ([
StructField("customer_id", IntegerType(), True),
StructField("name", StringType(), True),
StructField("city", StringType(), True),
StructField("state", StringType(), True),
StructField("country", StringType(), True),
StructField("registration_date", StringType(), True),
StructField("is_active", BooleanType(), True),
])
```

```python
# explicitly providing the schema; then no jobs ran this means it's a transformation
df = spark.read \
.format("csv") \
.option("header", "true") \
.schema(schema) \
.load("/data/customers_300mb.csv")
```
