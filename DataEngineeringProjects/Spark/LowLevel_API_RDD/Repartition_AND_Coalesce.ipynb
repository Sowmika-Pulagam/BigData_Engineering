```python
from pyspark.sql import SparkSession

#Intitalize Spark Session
spark = SparkSession.builder \
.appName ("Repartition_Vs_Coalesce") \
.getOrCreate ()
```

```python
!hadoop fs -ls /tmp/
```

```python
# Using 1mb file
rdd = spark.sparkContext.textFile('/tmp/customers_1mb.csv')
```

```python
rdd.take(5)
```

```python
rdd.getNumPartitions()
```

```python
## Repartition used when we want to increase/decrease the number of partitions.
## TO INCREASE THE PARTITION

rdd.repartition(4)
```

```python
reparition_rdd = rdd.repartition(4)
```

```python
reparition_rdd.getNumPartitions()
```

```python
rdd.repartition(4).getNumPartitions()
```

```python
## To decrease the partition from 4 to 1 partition
reparition_rdd_less = rdd.repartition(1)
```

```python
reparition_rdd_less.getNumPartitions()
```

```python
### Coalesce is used when we want to reduce the number of partition.
coalesce_rdd = rdd.coalesce(1)
```

```python
coalesce_rdd.getNumPartitions()
```

```python
## this is not changing the partition if we change 1 to 10 in coalesce
coalesce_rdd.coalesce(10).getNumPartitions()
```
