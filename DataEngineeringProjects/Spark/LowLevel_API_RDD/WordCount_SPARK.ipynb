```python
import spark
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("WordCount") \
    .master('yarn') \
    .getOrCreate()
```

```python
hdfs_path = '/tmp/inputhdfsdbz.txt'  # in HDFS file system
```

```python
## reading the text data file from hdfs file path to spark
rdd = spark.sparkContext.textFile(hdfs_path)
```

```python
rdd.collect()
```

```python
words = rdd.flatMap(lambda line : line.split(' '))
words.collect()
```

```python
word_map = words.map(lambda word: (word,1))
word_map.collect()
```

```python
result = word_map.reduceByKey(lambda a,b : a+b)
result.collect()
```

```python
spark.stop()
```
